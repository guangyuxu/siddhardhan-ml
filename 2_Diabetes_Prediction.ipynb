```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_dataset = pd.read_csv("https://www.dropbox.com/s/uh7o7uyeghqkhoy/diabetes.csv?dl=1")
print(diabetes_dataset.shape)
diabetes_dataset.head()
```

```python
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
X = diabetes_dataset.drop(columns='Outcome')
Y = diabetes_dataset['Outcome']
```

```python
scaler = StandardScaler()
scaler.fit(X)
standardScalerX = scaler.transform(X)
X = standardScalerX
X[0:4]
```

```python
X_train, X_nontrain, Y_train, Y_nontrain = train_test_split(X,Y, test_size = 0.2, random_state=2)
X_cv, X_test, Y_cv, Y_test = train_test_split(X_nontrain, Y_nontrain, test_size = 0.5, random_state=2)
print(f'{X_train.shape[0]} {Y_train.shape[0]} {X_cv.shape[0]} {Y_cv.shape[0]} {X_test.shape[0]} {Y_test.shape[0]} ')
```

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
Y_train_predict = classifier.predict(X_train)
accuracy_score(Y_train_predict, Y_train)
```

```python
accuracy_score(classifier.predict(X_cv), Y_cv)
```

```python
accuracy_score(classifier.predict(X_test), Y_test)
```
